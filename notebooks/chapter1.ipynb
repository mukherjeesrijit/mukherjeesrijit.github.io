## 1.1 What is a Tensor?

An $n$-dimensional array. For a matrix $\mathbf{A} \in \mathbb{R}^{m \times n}$:

$$\mathbf{A} = \begin{bmatrix} a_{11} & a_{12} \\ a_{21} & a_{22} \end{bmatrix}$$

# What is a Tensor?


```python
import torch as tr # library
```

## Scalar

A scalar is a 0D number. It is just a single number.

```python
s = tr.tensor([2.]) # scalar
print(s.item())
```

## Vector 

A vector is 1D number list. It is a list of any length of numbers.

```python
v = tr.tensor([2.,3.]) # scalar
print(v)
print("length:", len(v)) # number of elements in the vector
```

```python
print("length:", v.shape) # shape of a vector
```

```python
print("norm:", tr.norm(v))
```

```python
k = 3
print("vector", v)
print("scalar multiply", k*v)
```

* Shape of a vector
* Norm of a vector
* Unit vector 
* Scalar multiplication

* Add two vectors
* Dot product of two vectors
* Cosine similarity of two vectors

## Matrix

```python
m = tr.tensor([[1.,2.,3.],[2.,3.,4.]]) # matrix as a list of row vectors.
print(m.numpy())
```

```python
print(m.shape) # The number of row vectors = number of rows, 
# The number of element is each row vector = number of columns
```

* Add/Subtract two matrices
* Multiply a matrix by a scalar
* Multiplication of two matrices
* Hadamard Product of two matrices

```python
m1 = tr.tensor([[1.,2.,3.],[2.,3.,4.]]) # matrix 1
m2 = tr.tensor([[1.,2.,3.],[2.,3.,4.]]) # matrix 2
m3 = tr.tensor([[1.,2.,3.],[2.,3.,4.],[5.,6.,7.]]) # matrix 2

print(m1)
print(m1+m2) # matrix addition
print(m1-m2) # matrix addition
print(m1@m3) # matrix multiplication
print(m1*m2) # hadamard product
print(m1.T)
```

* M = A@B 
* M_ij = dot product of (ith row of A) and (jth column of B)
* row vectors of A should have the same length as column vectors of B
* number of columns of A = number of rows of B
* A (mxn) @ B (p,q) - then A @ B is possible iff n = p

### Matrix with Vector

```python
v1 = tr.tensor([1.,2.,3.]) # 1D vector (3)
v2 = tr.tensor([[1.,2.,3.]]) # Row vector - matrix with 1 row = vector (1,3)
v3 = tr.tensor([[1.],[2.],[3.]]) # Column vector - matrix with 1 column = vector (3,1)

# print(v1)
print(v1.shape)
# print(v2)
print(v2.shape)
# print(v3)
print(v3.shape)
print(v2.T.shape)
```

```python
m1 = tr.tensor([[1.,2.,3.],[2.,3.,4.]]) # matrix 1 = (2,3)
v1 = tr.tensor([1.,2.,3.]) # 1D vector (3)
v2 = tr.tensor([[1.,2.,3.]]) # Row vector - matrix with 1 row = (1,3)
v3 = tr.tensor([[1.],[2.],[3.]]) # Column vector - matrix with 1 column = (3,1)

print(m1@v1)
# print(m1@v2)
print(m1@v3)
```

```python
m1 = tr.tensor([[1.,2.,3.],[2.,3.,4.]]) # matrix 1 = (2,3)
v1 = tr.tensor([1.,2.,3.]) # 1D vector (3)
v3 = tr.tensor([[1.],[2.],[3.]]) # Column vector - matrix with 1 column = (3,1)

print((m1@v1).shape)
print((m1@v3).shape)

print(m1@v3+2.0) # broadcasting
```

### Tensors (nD)

```python
t = tr.ones([3,4,5,9,10,3,5,6])
print(t.shape)
```

```python
# broadcasting

t1 = tr.ones([4,6,1,3,6,1]) # 4,6,4,3,6,9
t2 = tr.ones([4,6,4,3,1,9]) # 4,6,4,3,6,9

# either they are equal or one of them has to be 1

print((t1+t2).shape)
```

```python
t1 = tr.ones([1])
print(t1)
print(t1.shape)
t2 = tr.ones([1,2])
print(t2)
print(t2.shape)
print(t1+t2)
```
