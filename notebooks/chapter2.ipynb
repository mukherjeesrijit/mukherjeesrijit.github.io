# Chapter 2: Autograd

## 2.1 Computational Graph

For $y = \mathbf{w}^\top \mathbf{x}$, gradient is:

$$\frac{\partial y}{\partial \mathbf{w}} = \mathbf{x}$$

## 2.2 Version Counter

Every tensor has `_version`. Inplace ops increment it.
If autograd saved the tensor before the inplace op → error on backward.


```python
import torch

x = torch.tensor([2.0], requires_grad=True)
y = x * x
print('version before inplace:', x._version)

# Safe: x not saved for backward by y
# Unsafe: inplace on tensor saved for grad
try:
    z = x * x
    x.data.add_(1)  # safe: .data bypasses autograd
    z.backward()
    print('grad:', x.grad)
except RuntimeError as e:
    print('Error:', e)
```
