# Introduction

In modern AI architectures, attention is an important architectural component for feature extraction. Every input signal $x$ is first represented a sequence of exhaustive of $n$ tokens $[x_1, x_2, \cdots, x_n]$ (row-wise like numpy/pytorch), where $x_i \in \mathbb{R}^d$ encodes both the signal and positional information of the token $i$. This makes $x$ a matrix of dimension $(n,d)$.
