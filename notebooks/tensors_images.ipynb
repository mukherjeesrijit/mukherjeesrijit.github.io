* A vector is 1D tensors
* A matrix is 2D tensors.
* What about nD tensors?

```python
import torch as tr
```

```python
v = tr.tensor([1.,2.,3.])
m = tr.tensor([[1.,2.,3.],[1.,2.,3.]])
rp = tr.tensor([[[1.,2.,3.],[1.,2.,3.]],[[1.,2.,3.],[1.,2.,3.]],[[1.,2.,3.],[1.,2.,3.]],[[1.,2.,3.],[1.,2.,3.]]])
```

```python
print(v.shape)
print(m.shape)
print(rp.shape)
```

```python
t =tr.ones([4,3,1,4,5,1,6])
print(t.shape)
```

```python
# Broadcasting (what are the rules of broadcasting)

t1 =tr.ones([4,3,1,4,5,9,6]) # [x0, x1, ..., xn]
t2 =tr.ones([4,3,3,4,5,1,6]) # [y0, y1, ..., yn]

print((t1*t2).shape)

# xi = yi or atleast of xi/yi should be 1.
```

```python
t1 = tr.ones([1,2]) # row vector -> extended along the other row 2 -> 2x2 matrix
print(t1)
t2 = tr.ones([2,1]) # column vector -> extended along the other column 2 -> 2x2 matrix
print(t2)
print(t1+t2)
```

```python
t = [[1.,1.,1.],[1.,1.,1.],[1.,1.,1.]]
```

```python
x = tr.tensor([[[1,2,3],
               [4,5,6],[5,6,7]],[[1,2,3],
               [4,5,6],[5,6,7]]]) # (2,3,3)
print(x.shape)
y = tr.sum(x, dim = 2, keepdims = True) # (2,3)
print(y.shape) 
```

```python
a = tr.tensor([[1,2],[3,4]]) # (2,2)
b = tr.tensor([[3],[4]]) # (2,1)
print(tr.cat([a,b], dim = 1).shape)
```

```python
x = tr.tensor([1,2,3])
x.unsqueeze(0).expand([4,3]).shape
```

```python
x = tr.tensor([1, 2, 3], dtype=tr.float32)
y = tr.tensor([4, 5, 6], dtype=tr.float32)
print(tr.outer(x,y))
```
